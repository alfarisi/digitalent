#Tugas List Nilai

Input : 
List nilai tugas yang berupa 250 angka random dengan rentang nilai 1-100 

Output :
* Angka ganjil
* Jumlah yang mendapat nilai > 70
* Standar deviasi
* Nilai maksimum dan minimum
* Modus

```python
import random

# generate nilai acak
nilaiTugas = []
for i in range(250):
    nilaiTugas.append(random.randrange(1,100,1))

print("List Nilai :", nilaiTugas)
angkaGanjil = []
nilai70plus = []
nilaiCount = {}
nilaiTmpList = []
frekTmpList = []
modusTmpDict = {}
total = 0
datalen = len(nilaiTugas)

for i,v in enumerate(nilaiTugas) :
    if v % 2 == 1 :
        angkaGanjil.append(v)
    
    if v > 70 :
        nilai70plus.append(v)
    
    total = total + v
    
    if v not in nilaiTmpList :
        nilaiCount[v] = 1
        nilaiTmpList.append(v)
    else :
        nilaiCount[v] = nilaiCount[v] + 1
    
    if nilaiCount[v] not in frekTmpList :
        frekTmpList.append(nilaiCount[v])
        modusTmpDict[nilaiCount[v]] = [v]
    else :
        modusTmpDict[nilaiCount[v]].append(v)
    
    if i == 0 :
        nilaiMin = v
        nilaiMax = v
        
    if v < nilaiMin :
        nilaiMin = v

    if v > nilaiMax :
        nilaiMax = v

mean = total / datalen
sum1 = 0

for x in nilaiTugas :
    sum1 = sum1 + (x - mean)**2

sd = (sum1 / (datalen - 1))**(0.5)
maxMuncul = 1

for i in frekTmpList :
    if i > maxMuncul :
        maxMuncul = i

print("\nAngka Ganjil :", angkaGanjil)
print("\nYang mendapat nilai > 70 ada", len(nilai70plus), "orang")
print("Standar Deviasi :", sd)
print("Nilai terkecil :", nilaiMin)
print("Nilai terbesar :", nilaiMax)

if maxMuncul == 1 :
    print("Modus : tidak ada")
else :
    print("Modus : ", modusTmpDict[maxMuncul])
```
